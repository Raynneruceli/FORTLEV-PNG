```python
import folium
import random
import pandas as pd

# Crie um objeto de mapa
mapa = folium.Map(location=[-22.8766521181865, -43.2278751249952], zoom_start=8)

# Lista de coordenadas e nomes de cidades fictícias do DataFrame
coordenadas_cidades = [([-22.8766521181865, -43.2278751249952], 'Rio de Janeiro'),
                      ([-22.9035, -43.2096], 'Rio de Janeiro'),
                      ([-22.9292, -43.1754], 'Niterói'),
                      ([-22.9105, -43.1687], 'São Gonçalo'),
                      ([-22.9083, -43.1964], 'Itaboraí'),
                      ([-22.8688, -43.3092], 'Duque de Caxias'),
                      ([-22.7664, -43.3131], 'Nova Iguaçu'),
                      ([-22.7548, -43.4507], 'Queimados'),
                      ([-22.7548, -43.7157], 'Angra dos Reis'),
                      ([-22.9472, -43.1852], 'Maricá')]

# Crie um DataFrame com os dados das cidades
df_cidades = pd.DataFrame(coordenadas_cidades, columns=['Coordenadas', 'Cidade'])


# Crie um menu na lateral direita com a descrição de cada número e um botão de alternância
menu_html = '<div id="cityInfo" style="position: fixed; top: 10px; right: 10px; z-index: 1000; background-color: #f0f0f0; padding: 10px; border-radius: 5px; width: 200px;">'

# Adicione a imagem de logo
menu_html += '<img src="https://www.fortlev.com.br/wp-content/themes/fortlev/assets/downloads/sem_slogan_cor.png" alt="Logo" style="max-width: 100%; display: block; margin: 0 auto 10px;"/>'

menu_html += '<button id="toggleButton" class="modern-button" onclick="toggleMarkers()">Mostrar Números</button>'
menu_html += '<div id="cityDetails" style="width: 100%; overflow: auto; max-height: 500px;">'
for idx, (_, cidade) in df_cidades.iterrows():
    menu_html += f'<p id="city_{idx}" class="cityInfo" style="margin: 0; padding: 5px;"><strong>{idx + 1}:</strong> {cidade}</p>'
menu_html += '</div>'
menu_html += '</div>'
mapa.get_root().html.add_child(folium.Element(menu_html))

# Adicione estilos CSS personalizados para o botão
css = """
<style>
.modern-button {
    background-color: #00386D; /* Cor de fundo */
    color: white; /* Cor do texto */
    border: none;
    padding: 10px 20px;
    border-radius: 5px;
    cursor: pointer;
    font-size: 16px;
}

.modern-button:hover {
    background-color: #0056b3; /* Cor de fundo ao passar o mouse */
}
</style>
"""
mapa.get_root().html.add_child(folium.Element(css))

# Adicione marcadores personalizados com ícone e número da cidade
markers = []

for idx, (coordenada, _) in df_cidades.iterrows():
    # Substitua o ícone atual
    icon = folium.DivIcon(
        icon_size=(None, None),
        icon_anchor=(0, 0),
        html=f'<div id="marker_{idx}" class="cityMarker" style="font-size: 14pt; background-color: #00386d; color: white; border-radius: 50%; text-align: center; width: 30px; height: 30px; line-height: 30px;">{idx + 1}</div>'
    )

    marker = folium.Marker(coordenada, icon=icon)
    marker.add_to(mapa)
    markers.append(marker)

# Crie um dicionário de mapeamento de nomes de cidades para coordenadas
city_coordinates = {cidade: coordenada for coordenada, cidade in df_cidades.itertuples(index=False)}

# Adicione nomes das cidades como camada oculta
city_layer = folium.FeatureGroup(name='Nomes das Cidades', show=False)
for cidade, coordenada in city_coordinates.items():
    folium.Marker(
        location=coordenada,
        icon=folium.DivIcon(html=f'<div class="cityLabel">{cidade}</div>', icon_size=(None, None), icon_anchor=(0, 0)),
        tooltip=cidade  # Adicione um tooltip para mostrar o nome da cidade ao passar o mouse
    ).add_to(city_layer)
city_layer.add_to(mapa)

# Adicione estilos CSS personalizados para os marcadores de nomes de cidades
css_city_label = """
<style>
.cityLabel {
    background-color: rgba(255, 255, 0, 0.9); /* Fundo amarelo com 70% de transparência */
    color: black; /* Cor do texto */
    border: 1px solid gray; /* Borda cinza */
    border-radius: 2px;
    padding: 1px;
    font-size: 12px; /* Tamanho do texto */
    text-align: center;
    white-space: nowrap; /* Evita quebra de linha */
}
</style>
"""
mapa.get_root().html.add_child(folium.Element(css_city_label))

# JavaScript para alternar entre números e nomes
javascript = """
<script>
var showNumbers = true;

function toggleMarkers() {
    var cityMarkers = document.getElementsByClassName("cityMarker");
    var cityLabels = document.getElementsByClassName("cityLabel");

    if (showNumbers) {
        showNumbers = false;
        document.getElementById("toggleButton").textContent = "Mostrar Números";

        // Ocultar os números no mapa
        for (var i = 0; i < cityMarkers.length; i++) {
            cityMarkers[i].style.display = "none";
        }

        // Mostrar os nomes das cidades no mapa
        for (var i = 0; i < cityLabels.length; i++) {
            cityLabels[i].style.display = "block";
        }
    } else {
        showNumbers = true;
        document.getElementById("toggleButton").textContent = "Mostrar Nomes";

        // Mostrar os números no mapa
        for (var i = 0; i < cityMarkers.length; i++) {
            cityMarkers[i].style.display = "block";
        }

        // Ocultar os nomes das cidades no mapa
        for (var i = 0; i < cityLabels.length; i++) {
            cityLabels[i].style.display = "none";
        }
    }
}
</script>
"""
mapa.get_root().html.add_child(folium.Element(javascript))

# Salvar o mapa em um arquivo HTML
mapa.save('mapa_com_botao_e_descricao_com_logo.html')
```
